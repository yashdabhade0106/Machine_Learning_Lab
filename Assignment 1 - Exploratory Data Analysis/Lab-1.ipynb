```python
data = data.drop(['S.No.'], axis=1)
data.info()
```

```python
from datetime import date
date.today().year
data['Year'] = date.today().year - data['Year']
data.head
```

```python
data['Brand'] = data.Name.str.split().str.get(0)
```

```python
data["Model"] = data.Name.str.split().str.get(1) + data.Name.str.split().str.get(2)
data.head()
```

```python
data[['Name','Brand','Model']]
```

```python
data['Average_Kilometers'] = data['Kilometers_Driven']
```

```python
from datetime import date
date.today().year
data['Car_Age']=date.today().year-data['Year']
data.head()
```

```python
data['KM_Avg_per_Year'] = data['Kilometers_Driven'] / data['Car_Age']
data.head()
```

```python
print(data.Brand.unique())
print(data.Brand.nunique())
```

```python
searchfor = ['Isuzu' ,'ISUZU','Mini','Land']
data[data.Brand.str.contains('|'.join(searchfor))].head(5)
```

```python
data["Brand"].replace({"ISUZU": "Isuzu", "Mini": "Mini Cooper","Land":"Land Rover"}, inplace=True)
```

```python
print(data.Brand.unique())
print(data.Brand.nunique())
```

```python
data.head()   
```

```python
data['Is_Recent_Car'] = data['Car_Age'] <= 5
recent_cars = data[data['Is_Recent_Car']]
print(data.head())
print(recent_cars.head()) 
```

```python
data['Fuel_Type'].value_counts()

```

```python
import matplotlib.pyplot as plt
lables = data['Fuel_Type'].value_counts().index
sizes = data['Fuel_Type'].value_counts().values

plt.bar(lables, sizes)
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.savefig('fuel_type_distribution.png')
plt.title('Fuel Type Distribution of Used Cars')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
lables = data['Fuel_Type'].value_counts().index
sizes = data['Fuel_Type'].value_counts().values
colors = ['orange','blue','red','pink']
plt.pie(sizes, labels=lables, autopct='%1.1f%%', startangle=100)
plt.axis('equal')  
plt.savefig('fuel_type_distribution.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
lables = data['Brand'].value_counts().head(10).index
sizes = data['Brand'].value_counts().head(10).values
colors = ['orange','blue','red','pink']
plt.pie(sizes, labels=lables, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  
plt.savefig('brand_distribution.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.hist(data['Price'], bins=20, color='blue', edgecolor='black')
plt.xlabel('Price') 
plt.ylabel('Frequency')
plt.title('Price Distribution of Used Cars')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.hist(data['Year'], bins=20, color='blue', edgecolor='black')
plt.xlabel('Year') 
plt.ylabel('Numbers Of Cars')
plt.title('Year Distribution of Used Cars')
plt.savefig('year_distribution.png')
plt.grid()
plt.show()
```

```python
fig,axxr = plt.subplots(1, 2, figsize=(12, 5))
data.groupby('Brand')['Price'].mean().plot(kind='bar', ax=axxr[0], color='#4ECDC4')
axxr[0].set_title('Average Price by Car Type')
data.groupby('Brand')['Seats'].mean().plot(kind='bar', ax=axxr[1], color='#FF6B6B')
axxr[1].set_title('Average Seats by Car Type')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['Year'], data['Price'], alpha=0.5)
plt.xlabel('Year')  
plt.ylabel('Price')
plt.title('Price vs Year of Used Cars')
plt.savefig('price_vs_year.png')
plt.grid()
plt.show()
```
